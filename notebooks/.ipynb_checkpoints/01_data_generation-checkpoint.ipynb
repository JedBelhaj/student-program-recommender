```python
import numpy as np
import pandas as pd
import random
```

```python
programs = [
    ("Computer Science", "Programming, algorithms, AI", "programming algorithms math ai technology"),
    ("Biology", "Life sciences and research", "biology genetics research science ecology life"),
    ("Landscape Architecture", "Ecology, design, outdoor planning", "biology design drawing ecology"),
    ("Business Analytics", "Data-driven business decisions", "data statistics business technology"),
    ("Fine Arts", "Creative visual expression", "drawing painting creativity"),
    ("Environmental Engineering", "Sustainability and engineering", "environment engineering ecology technology"),
    ("Psychology", "Human behavior and mental processes", "psychology science research data behavior"),
    ("Graphic Design", "Visual communication and branding", "design art drawing creativity technology"),
    ("Data Science", "Analytics and machine learning", "data programming math statistics ai technology"),
    ("Architecture", "Building design and urban planning", "design drawing art engineering creativity"),
    ("Economics", "Markets and financial systems", "business math data statistics research"),
    ("Mechanical Engineering", "Machines and systems design", "engineering math technology design"),
    ("Marketing", "Brand strategy and consumer insights", "business creativity data design"),
    ("Literature", "Writing and literary analysis", "writing research creativity art"),
    ("Chemistry", "Matter and chemical reactions", "science research math biology technology"),
]

programs_df = pd.DataFrame(programs, columns=["name", "description", "tags_text"])
programs_df["program_id"] = ["p_" + str(i) for i in range(len(programs_df))]
```

```python
interests_pool = [
    "biology", "design", "drawing", "math", "programming",
    "ecology", "art", "data", "business", "engineering",
    "psychology", "statistics", "writing", "science", "creativity"
]

users = []
for i in range(100):  # 100 users is perfect
    interests = " ".join(random.sample(interests_pool, 3))
    grade = round(np.random.normal(13, 2), 1)  # realistic grades
    users.append((f"u_{i}", interests, grade))

users_df = pd.DataFrame(users, columns=["user_id", "interests_text", "avg_grade"])
```

```python
interactions = []

for _, user in users_df.iterrows():
    # Each user interacts with 3-6 programs (more realistic)
    num_interactions = random.randint(3, 6)
    liked_programs = random.sample(list(programs_df["program_id"]), num_interactions)
    for p in liked_programs:
        interactions.append((user["user_id"], p, 1))

interactions_df = pd.DataFrame(interactions, columns=["user_id", "program_id", "interaction"])
```

```python
users_df.to_csv("../data/raw/users.csv", index=False)
programs_df.to_csv("../data/raw/programs.csv", index=False)
interactions_df.to_csv("../data/raw/interactions.csv", index=False)
```
