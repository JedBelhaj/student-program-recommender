```python
import numpy as np
import pandas as pd
import random
```

```python
programs = [
    ("Computer Science", "Programming, algorithms, AI", "programming algorithms math ai"),
    ("Biology", "Life sciences and research", "biology genetics research"),
    ("Landscape Architecture", "Ecology, design, outdoor planning", "biology design drawing ecology"),
    ("Business Analytics", "Data-driven business decisions", "data statistics business"),
    ("Fine Arts", "Creative visual expression", "drawing painting creativity"),
    ("Environmental Engineering", "Sustainability and engineering", "environment engineering ecology"),
]

programs_df = pd.DataFrame(programs, columns=["name", "description", "tags_text"])
programs_df["program_id"] = ["p_" + str(i) for i in range(len(programs_df))]
```

```python
interests_pool = [
    "biology", "design", "drawing", "math", "programming",
    "ecology", "art", "data", "business", "engineering"
]

users = []
for i in range(100):  # 100 users is perfect
    interests = " ".join(random.sample(interests_pool, 3))
    grade = round(np.random.normal(13, 2), 1)  # realistic grades
    users.append((f"u_{i}", interests, grade))

users_df = pd.DataFrame(users, columns=["user_id", "interests_text", "avg_grade"])
```

```python
interactions = []

for _, user in users_df.iterrows():
    liked_programs = random.sample(list(programs_df["program_id"]), 2)
    for p in liked_programs:
        interactions.append((user["user_id"], p, 1))

interactions_df = pd.DataFrame(interactions, columns=["user_id", "program_id", "interaction"])
```

```python
users_df.to_csv("../data/raw/users.csv", index=False)
programs_df.to_csv("../data/raw/programs.csv", index=False)
interactions_df.to_csv("../data/raw/interactions.csv", index=False)
```
